```python
import string  
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn import preprocessing 
import re
```

```python
# veri seti okumasi
data=pd.read_csv("Twitter_Data.csv")
```

```python
data=data.dropna()
```

```python
data['Tweet'] = data['Tweet'].str.replace('\d+', '')
data['Tweet'] = data['Tweet'].str.rstrip(string.digits)
```

```python
data[data.Etiket=='mutlu'].count()
data[data.Etiket=='korku'].count()
data[data.Etiket=='kızgın'].count()
data[data.Etiket=='surpriz'].count()
data[data.Etiket=='üzgün'].count()
```

```python
label_encoder =preprocessing .LabelEncoder()
```

```python
data['Etiket']= label_encoder.fit_transform(data['Etiket'])
```

```python
# X=data.Tweet.string.lower()
X=[str(x.lower()) for x in data.Tweet]
y=data.Etiket
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2,shuffle=True ,test_size=0.2)
```

```python
X_train=[x for x in X_train]
```

```python
X_test=[x for x in X_test]
```

```python
# tokenizer=Tokenizer(oov_token=
#
# tokenizer.fit_on_texts(X_train)
# word_index=tokenizer.word_index
```

```python
oov_tok = "<OOV>"
tokenizer = Tokenizer(num_words=10000, oov_token=oov_tok)
tokenizer.fit_on_texts(X_train)

word_index = tokenizer.word_index

training_sequences = tokenizer.texts_to_sequences(X_train)
training_padded = pad_sequences(training_sequences, maxlen=100, padding='post', truncating='post')

testing_sequences = tokenizer.texts_to_sequences(X_test)
testing_padded = pad_sequences(testing_sequences, maxlen=100, padding='post', truncating='post')
```

```python

training_padded = np.array(training_padded)
# training_labels = np.array(y_train)
testing_padded = np.array(testing_padded)
# testing_labels = np.array(y_test)
```

```python
y_train=np.array(y_train)
print(type(y_train))
print(y_train)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(10000, 16, input_length=100),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(24, activation='relu'),
     tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(5, activation='sigmoid')
])
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(training_padded, y_train, epochs=20, verbose=2)
```

```python
import matplotlib.pyplot as plt


def plot_graphs(history, string):
  plt.plot(history.history[string])
  plt.plot(history.history[string])
  plt.xlabel("Epochs")
  plt.ylabel(string)
  plt.legend([string, string])
  plt.show()
plot_graphs(history, "accuracy")
plot_graphs(history, "loss")
```

```python
y_test=np.array(y_test)
```

```python
histo=model.evaluate(testing_padded,y_test)
```
